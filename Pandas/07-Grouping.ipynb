### <center>Grouping</center>

```python
df = pd.read_csv("C:/Users/Lenovo/Desktop/Python/Data Science/Pandas/Extra files/cars_dataset.csv")
df
```

```python
a = df.groupby(by='model_year')
a
```

Notice that what is returned is not a set of DataFrames, but a DataFrameGroupBy object. This object is where the magic is: you can think of it as a special view of the DataFrame, which is poised to dig into the groups but does no actual computation until the aggregation is applied.

```python
df.groupby(by='model_year').mean()["mpg"]
```

#### <center>Aggregate, filter, transform </center>
The preceding discussion focused on aggregation for the combine operation, but there are more options available. In particular, GroupBy objects have aggregate(), filter(), transform(), and apply() methods that efficiently implement a variety of useful operations before combining the grouped data.

##### Aggregation
We’re now familiar with GroupBy aggregations with sum(), median(), and the like, but the aggregate() method allows for even more flexibility. It can take a string, a function, or a list thereof, and compute all the aggregates at once.

```python
agg_values = a.agg({"acceleration": np.mean,
                          "displacement": [min, max]}) # .aggregate() can also be used.
print(agg_values)
```

#### Filtering.
A filtering operation allows you to drop data based on the group properties.


```python
def filt_fun(x):
    return x["displacement"].min() > 80


filter_fun = a.filter(filt_fun)
filter_fun
```

### <center>Multi-level Index</center>

```python
# Higer level grouping
df.groupby(by=["model_year", 'cylinders']).mean()
```

```python
# Multi-level Index
df.groupby(by=["model_year", 'cylinders']).mean().index
```

```python
year_cyl = df.groupby(by=["model_year", 'cylinders']).mean()
year_cyl.index.levels
```

```python
year_cyl.loc[70] # only works on the outside index
```

```python
year_cyl.loc[(70, 4)] # For the inner level, it's importat to use the tuple
```

```python
# Cross-section --> takes a key, axis and a level
year_cyl.xs(key=70, level='model_year') # Cross section only takes one key, not multiple
```

```python
year_cyl.xs(key=4, level="cylinders")
```

```python
# year_cyl.swaplevel() is used to swap the levels
```

```python
# Sorting

year_cyl.sort_index(level="model_year", ascending=False) # In this case, it is preferred to order the outerlevel
```

#### Example

```python
index = pd.MultiIndex.from_product([[2013, 2014], [1, 2]], names=['year', 'visit'])
columns = pd.MultiIndex.from_product([['Bob', 'Guido', 'Sue'], ['HR', 'Temp']], names=['subject', 'type'])
# In this case a Data Frame of 4 dimensions is created and represented in a 4x6 matrix, names of the columns and indexes
# give a better understanding in the distribution of the matrix

data = np.round(np.random.randn(4, 6), 1)
data[:, ::2] *= 10
data += 37

df_ex = pd.DataFrame(data, index=index, columns=columns)
df_ex
```

```python
#  Data Frame slicing
idx_slc = pd.IndexSlice
alfa = df_ex.loc[idx_slc[:, 1], idx_slc["Bob", "HR"]] # Advanced slicing, in this case the slices in Index and 
                                                      # columns can be chosen
alfa
```

```python
# Perhaps we’d like to average out the measurements in the two visits each year. We can do this by naming the index
# level we’d like to explore, in this case the year. By calling level=year Python will take the average number between
# the 2 years, same would have been done if type, visit or any other column/axis is selected.

data_mean = df_ex.mean(level='year')
data_mean
```
