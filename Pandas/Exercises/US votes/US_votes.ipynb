```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('primary_results.csv')
df
```

```python
# To access the first or last lines .head() or .tail() are used

firsts = df.head(3)
firsts
```

```python
lasts = df.tail(3)
lasts
#                                       To display all the columns and rows:
# pd.set_option('display.max_columns', #)
# pd.set_option('display.max_rows', #)
```

```python
df = df.set_index("state")  # In this case state becomes the index
# To reset the index use: votes.reset_index()
# To sort the index use votes.sort_index()
```

```python
# In this case a data frame is printed out with all the "indexes" or values
df.loc["Texas"].head()
# referred to Texas
```

```python
# A series is printed out with the indexes as states and values as candidate votes
df["candidate"].tail()
```

```python
# To have numerical data about any column use .value_counts()
# in this case it takes as the index values of the column
total_candidate_counts = df["candidate"].value_counts()
total_candidate_counts
```

```python
# Filtering - Using Conditionals to Filter Rows and Columns
# in this case to perform a filter, a statement is declared with a condition to be filtered
voted_zero = df.votes == 0
# declaration is evaluated in the df (evaluation can be performed in the declaration itself)
df[voted_zero]
```

```python
votes_NY = df[(df.county == "Manhattan") & (
    df.party == "Democrat")]  # and statement
votes_NY
```

```python
votes_NY_vs_LA = df[(df.county == "Manhattan") | (
    df.county == "Los Angeles")]  # or statement
votes_NY_vs_LA
```

```python
states = ["CA", "TX"]
# In this type of filter .isin() a logical value is printed
logical_filter = df['state_abbreviation'].isin(states)
# if True, value's printed out, in this case an extra filter is
result = df.loc[logical_filter, "fraction_votes"]
result
```

```python
# searches with a string method the word and outputs a boolean
word_filter = df["candidate"].str.contains("Trump")
# if True, value's printed out, in this case an extra filter is
filtered_data = df.loc[word_filter, "fraction_votes"]
filtered_data  # applied where only the fraction votes are printed out
```

```python
#                                              Sorting
# Values
votes_sorted = df.sort_values(by="votes", ascending=False)
votes_sorted
```

```python
# Data frames
filter_sort = df.sort_values(by=["votes", "fraction_votes"], ascending=[False, False])
filter_sort
```

```python
#                           Grouping and Aggregating - Analyzing and Exploring Data

# For a general analysis the .describe() is used
print(df.describe())  # if it´s numerical data it makes statistics
df['party'].describe()  # if it isn't, general data is provided
```

```python
states_grouped = df.groupby("state")  # In this case a new df is created where all the values are merged by state,
# however it can't be printed

total_votes_state = states_grouped.votes.sum()
total_votes_state
```

```python
print(f"The total sum of votes in the US election was {total_votes_state.sum()} votes")
```

```python
candidate_state = states_grouped["candidate"].value_counts().loc["Texas"]
candidate_state
```

```python
# Many series can be grouped
votes_party_state = df.groupby(["state", "party"]).votes.sum()
votes_party_state
```

```python
# .agg() helps to use more than one statistical formula
mean_and_median = states_grouped["votes"].agg(['mean', 'median'])
mean_and_median
```

```python
state_appearance = df["state_abbreviation"].value_counts()  # Series with values of appearance of each state

group_s = df.groupby("state_abbreviation")

trump_supporters = group_s["candidate"].apply(lambda x: x.str.contains("Trump").sum())  # In this case a function is
# applied to candidates searching for Trump, if True value is summed

trump_df = pd.concat([state_appearance, trump_supporters], axis=1, sort=False)
# axis = 0 concatenates values by stacking them (rows) --> creates a series
# axis = 1 concatenates values by placing them together (columns) --> creates a data frame
trump_df.rename(columns={'state_abbreviation': 'total_votes', 'candidate': 'Trump_supporters'}, inplace=True)
trump_df
```

```python

trump_df["percentage"] = 100 * (trump_df["Trump_supporters"]) / (trump_df['total_votes'])
trump_df

# Duplicates indexes: to verify that the indices in the result of pd.concat() do not overlap, specify the
# verify_integrity flag set to True (use it in a try except loop).
# Ignore index: useful in pd.concat() to reindex elements --> ignore_index=True
```

```python
#                                       Merging data frames

# To merge two data frames a common set of values (column) is needed, without this merging is impossible

poverty_df = pd.read_csv('PovertyEstimates.csv')
poverty_df.head()
```

```python
# To merge two datasets with different column names it is used the left_on and right_on keywords to specify the two
# column names:

merged_df = df.merge(poverty_df, left_on="fips", right_on="FIPStxt")
merged_df.drop(columns=["FIPStxt"], inplace=True)
merged_df.rename(columns={"PCTPOVALL_2015": "poverty_estimate_2015"}, inplace=True)
merged_df
```

```python
merged_df.to_csv("new_df.csv")
```
