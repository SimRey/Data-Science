## AirBnB exercise

```python
filt = (df['accommodates'] >= 4) & (df['bedrooms'] == 2) & (df['reviews'] > 10) & (df['overall_satisfaction'] >= 4)

filtered_data = df[filt]
sorted_data = filtered_data.sort_values(by=["overall_satisfaction", 'reviews'], ascending=[False, False])
sorted_data.head()
```

## Crimes exercise

```python
data_frame = pd.read_excel("crimes_2019.xlsx")
data_frame.set_index("mes", inplace=True)
data_frame
```

```python
# a) Vehicles complaints
total_vehicle = (data_frame["Robo_total_de_vehículo"] + data_frame["Robo_parcial_de_vehículo"]) * 100 \
                / (data_frame["total"])
total_vehicle
```

```python
# b) Months with most house/street robberies
rob_house = data_frame["Robo_en_casa_habitación"].nlargest(3)
rob_house
```

```python
# c) Recurring crime
df = data_frame
df.drop(columns="total", inplace=True)
for i in df.index:
    max_value = df.loc[i].max()
    for j in df.columns:
        if df.loc[i,j] == max_value:
            print(f"{i} {j}: {max_value}")
            break
```

```python
# d) Numbers of each crime
for crime in df.columns:
    total = sum(df.loc[:,crime])
    print(f"{crime}:{total}")
```

```python
# e) mean, max and min for each crime
for c in df.columns:
    print(df[c].describe())
    print()
```
