
##### TASK 1: Scatter Plot

The scatterplot attempts to show the relationship between the days employed versus the age of the person (DAYS_BIRTH) for people who were not unemployed. Note, to reproduce this chart you must remove unemployed people from the dataset first. Also note the sign of the axis, they are both transformed to be positive.  


```python
filt = df["DAYS_EMPLOYED"] < 0

df1 = df[filt][["DAYS_EMPLOYED", "DAYS_BIRTH"]]

df1["DAYS_BIRTH"] = -1*df1["DAYS_BIRTH"] 
df1["DAYS_EMPLOYED"] = -1*df1["DAYS_EMPLOYED"] 

sns.scatterplot(data=df1, x="DAYS_BIRTH", y="DAYS_EMPLOYED")


```

#### TASK 3: Categorical Plot

This plot shows information only for the *bottom half* of income earners in the data set. It shows the boxplots for each category of NAME_FAMILY_STATUS column for displaying their distribution of their total income. Note: You will need to adjust or only take part of the dataframe *before* recreating this plot. You may want to explore the *order* parameter to get the xticks in the exact order shown here

```python
filt2 = df["AMT_INCOME_TOTAL"] < df["AMT_INCOME_TOTAL"].mean()
df2 = df[filt2][["AMT_INCOME_TOTAL", "NAME_FAMILY_STATUS", "FLAG_OWN_REALTY"]]

plt.figure(figsize=(12,6))
sns.boxplot(data=df2, x="NAME_FAMILY_STATUS", y="AMT_INCOME_TOTAL", hue="FLAG_OWN_REALTY")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., title='FLAG_OWN_REALTY')
```

#### TASK 4: Heat Map

This heatmap shows the correlation between the columns in the dataframe. You can get correlation with .corr() , also note that the FLAG_MOBIL column has NaN correlation with every other column, so you should drop it before calling .corr().

```python
a = df.dropna(axis='index', how='any') 
a.drop(columns="FLAG_MOBIL", inplace=True)

df3 = a.corr()

sns.heatmap(df3)
```
